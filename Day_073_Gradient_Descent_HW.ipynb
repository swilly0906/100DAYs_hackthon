使用function y=(x+5)² 來練習學習率( leanrning rate ) 調整所造成的影響

# 作業目標:
    請嘗試使用不同的組合驗證 learning rate 對所需 iteration 的影響
    lr = [0.1, 0.0001]
    主要驗證 Lr 對於grandient 收斂的速度
    

```python
cur_x = 3 # The algorithm starts at x=3
precision = 0.000001 #This tells us when to stop the algorithm
max_iters = 10000 # maximum number of iterations
iters = 0
df = lambda x: 2*(x+5) #Gradient of our function 
lr = [0.1, 0.0001]
previous_step_size = 1
x_history = [[cur_x],[cur_x]]
iters_history = [[iters],[iters]]
```

```python
'''
需要執行初始值設定, 下列三個
# Learning rate
#iteration counter
#Step size  
'''
'''
while previous_step_size > precision and iters < max_iters:

  算法迭代更新

print("the learning rate is",lr, "\nIteration",iters,"\nX value is",cur_x) #Print iterations
'''
```

```python
while (previous_step_size > precision) and (iters < max_iters):
    prev_x = cur_x
    cur_x = cur_x-lr[0]*df(cur_x)
    previous_step_size = abs(cur_x-prev_x)
    iters = iters+1
    print("iteration : ",iters, "\n x is ", cur_x)
    
    x_history[0].append(cur_x)
    iters_history[0].append(iters)
```

```python
import matplotlib.pyplot as plt
plt.plot(iters_history[0],x_history[0], "o-", color = "red")
```

```python
while (previous_step_size > precision) and (iters < max_iters):
    prev_x = cur_x
    cur_x = cur_x-lr[1]*df(cur_x)
    previous_step_size = abs(cur_x-prev_x)
    iters = iters+1
    print("iteration : ",iters, "\n x is ", cur_x)
    
    x_history[1].append(cur_x)
    iters_history[1].append(iters)
```

```python
plt.plot(iters_history[1],x_history[1], "o-", color = "red")
```
